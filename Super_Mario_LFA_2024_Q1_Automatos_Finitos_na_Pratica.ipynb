```python
def afd(M, w):
  # 'extrai' os elementos da tupla
  δ, q, F = M

  for s in w:
    q = δ[(q, s)]

  return q in F
```

```python
# funcao de transicao
δ = {
    ('q1', '0'): 'q1',
    ('q1', '1'): 'q2',
    ('q2', '1'): 'q2',
    ('q2', '0'): 'q3',
    ('q3', '0'): 'q2',
    ('q3', '1'): 'q2'}
```

```python
# AFD representado por uma 3-upla, ao invés de uma 5-upla
# O alfabeto e conjunto de estados podem ser inferidos da funcao de transicao
M = (δ, 'q1', ['q2'])
```

```python
afd(M, "000111101010000")
```
