# **Testes de diferentes sequ√™ncias de comandos de entrada para avaliar se o jogador conseguiria passar de fase**


---

Com o aut√¥mato finito determin√≠stico (AFD) gerado na etapa anterior, qualquer sequ√™ncia de entrada com comandos do jogador pode ser testada para verificar se a sequ√™ncia levaria o Mario a alcan√ßar a bandeira e vencer a fase. Al√©m disso, √© poss√≠vel analisar o motivo de determinadas sequ√™ncias n√£o terem sucesso.

Nos comandos abaixo, caso uma sequ√™ncia seja validada com "True", isso significa que os comandos utilizados levam o Mario a passar de fase. Caso a sequ√™ncia seja negada com "False", os comandos n√£o podem ser utilizados para fazer o Mario passar de fase, seja pelo Mario n√£o ter alcan√ßado a bandeira com a sequ√™ncia de comandos (4¬∞ teste) ou por ter morrido no percurso (3¬∞ teste).

```python
#Mario vence a fase sem utilizar o atalho
automata.accept("66969999999669669699966996996699996699696696969969")
```

```python
#Mario vence a fase utilizando o atalho pelo cano
automata.accept("66969999999296669669699669")
```

```python
#Mario morre no percurso
automata.accept("666")
```

```python
#Mario n√£o alcan√ßa a bandeira com apenas esses comandos
automata.accept("6696999")
```
