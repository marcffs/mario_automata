# **Testes de diferentes sequências de comandos de entrada para avaliar se o jogador conseguiria passar de fase**


---

Com o autômato finito determinístico (AFD) gerado na etapa anterior, qualquer sequência de entrada com comandos do jogador pode ser testada para verificar se a sequência levaria o Mario a alcançar a bandeira e vencer a fase. Além disso, é possível analisar o motivo de determinadas sequências não terem sucesso.

Nos comandos abaixo, caso uma sequência seja validada com "True", isso significa que os comandos utilizados levam o Mario a passar de fase. Caso a sequência seja negada com "False", os comandos não podem ser utilizados para fazer o Mario passar de fase, seja pelo Mario não ter alcançado a bandeira com a sequência de comandos (4° teste) ou por ter morrido no percurso (3° teste).

```python
#Mario vence a fase sem utilizar o atalho
automata.accept("66969999999669669699966996996699996699696696969969")
```

```python
#Mario vence a fase utilizando o atalho pelo cano
automata.accept("66969999999296669669699669")
```

```python
#Mario morre no percurso
automata.accept("666")
```

```python
#Mario não alcança a bandeira com apenas esses comandos
automata.accept("6696999")
```
